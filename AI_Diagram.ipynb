```python
import matplotlib.pyplot as plt
fish = ["Lele", "Gurame" ,"Nila"]
datamatch = [10, 20, 30]
colors = ["yellow", "lightskyblue", "gold"]
ax = plt.axes()
plt.pie(datamatch, labels = fish, colors = colors,  autopct="%1.0f%%", shadow=True, startangle=90)
ax.set_title("Fish Order Data")
plt.show()
```

**Bar Visual AI** ðŸ‘‡

```python
import matplotlib.pyplot as plt
fish = ["Lele", "Gurame" ,"Nila"]
datamatch = [10, 20, 30]
plt.figure(figsize=(10,6))
plt.bar(fish, datamatch, color= 'green')
plt.title("Fish Order Data", size=17)
plt.ylabel("Data Match", size=13)
plt.xlabel("Fish Name", size=13)
plt.xticks(size=11)
plt.yticks(size=11)
plt.show()
```
