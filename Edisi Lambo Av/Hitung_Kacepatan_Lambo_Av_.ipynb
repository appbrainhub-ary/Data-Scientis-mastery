Di bawah ini adalah implementasi simulasi kecepatan transmisi Lamborghini Aventador dengan perhitungan interval sederhana

```python
import math

# Konstanta & Algoritma
kepadatan_udara = 1.225
luas_depan = 2.23
koefisien_drag = 0.33
massa_kendaraan = 1575
daya_mesin = 566
rasio_gigi = [2.56, 1.81, 1.41, 1.13, 0.91, 0.75]
rasio_gandar_akhir = 3.70
jari_jari_ban = 0.335

# Parameter simulasi
waktu_simulasi = 10  # Waktu simulasi dalam detik
interval_waktu = 0.1  # Interval waktu antara perhitungan dalam detik

# Menghitung kecepatan maksimum
def hitung_kecepatan_maksimum():
    kecepatan_maksimum = 0

    for gigi in rasio_gigi:
        kecepatan = (daya_mesin * gigi * rasio_gandar_akhir * jari_jari_ban * math.pi) / (massa_kendaraan * interval_waktu)
        if kecepatan > kecepatan_maksimum:
            kecepatan_maksimum = kecepatan

    return kecepatan_maksimum

# Menjalankan simulasi
def jalankan_simulasi():
    kecepatan_maksimum = hitung_kecepatan_maksimum()

    waktu = 0
    posisi = 0
    kecepatan = 0

    while waktu <= waktu_simulasi:
        percepatan = (daya_mesin * rasio_gigi[0] * rasio_gandar_akhir * jari_jari_ban * math.pi) / (massa_kendaraan * interval_waktu)
        gaya_drag = 0.5 * kepadatan_udara * luas_depan * koefisien_drag * kecepatan ** 2
        gaya_bersih = percepatan - gaya_drag
        kecepatan += gaya_bersih / massa_kendaraan * interval_waktu
        posisi += kecepatan * interval_waktu

        print(f"Waktu: {round(waktu, 1)} detik, Kecepatan: {round(kecepatan * 3.6, 1)} km/jam, Posisi: {round(posisi, 1)} meter")

        waktu += interval_waktu

        # Batasi kecepatan menjadi kecepatan maksimum
        if kecepatan > kecepatan_maksimum:
            break

# Menjalankan simulasi
jalankan_simulasi()
```
