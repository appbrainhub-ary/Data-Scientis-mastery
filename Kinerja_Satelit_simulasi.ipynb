```python

import numpy as np
import matplotlib.pyplot as plt

print("üõ∞Ô∏è Simulasi Kinerja Satelit üõ∞Ô∏è")
print("üë®üèª‚ÄçüöÄ bot pilot: Ari")

# data orbit pesawat satelit

data_orbit = np.array([[0, 0], [1, 1], [2, 4], [3, 9], [4, 16]])

# pisahkan data ke x dan y

x = data_orbit[:, 0]
y = data_orbit[:, 1]

# hitung kecepatan satelit

velocity = np.diff(y) / np.diff(x)

# tampilkan grafik orbit dan kecepatan satelit

plt.plot(x, y, label='Orbit')
plt.plot(x[:-1], velocity, label='Kecepatan')
plt.xlabel('Waktu')
plt.ylabel('Parameter')
plt.legend()
plt.show()
```
