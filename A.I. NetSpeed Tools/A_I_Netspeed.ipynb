# **Netspeed A.I. (Tes Kecepatan Jaringan)**

**seperti hal nya tools yang populer seperti speedtest, di sini saya akan membuat extensi tambahan A.I. untuk cek / tes kecepatan bandwith jaringan per Mbps.**

```python

import speedtest

print("||| Extensi A.I. ||||")
print("Sistem Tools NetSpeed")
print("|||||| By.Ari |||||||")
print("result: ")

# Algoritmanya:

def test_kecepatan():
    st = speedtest.Speedtest()
    download_speed = st.download() / 10**6
    upload_speed = st.upload() / 10**6

    return download_speed, upload_speed

def peringatan_kecepatan(download_speed, upload_speed, standar_kecepatan):
    if download_speed < standar_kecepatan or upload_speed < standar_kecepatan:
        return " Network di bawah standar."
    else:
        return " Network memenuhi standar."

standar_kecepatan = 10  # Misalnya, standar kecepatan adalah 10 Mbps
download_speed, upload_speed = test_kecepatan()
peringatan = peringatan_kecepatan(download_speed, upload_speed, standar_kecepatan)

print("Down: {:.2f} Mbps".format(download_speed))
print("Up: {:.2f} Mbps".format(upload_speed))
print(peringatan)
```
