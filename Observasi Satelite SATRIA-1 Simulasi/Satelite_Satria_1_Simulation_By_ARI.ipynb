SATRIA-1 merupakan satelit super canggih pertama yang sepenuhnya dimiliki dan dikendalikan oleh Pemerintah Indonesia. Satelit buatan Thales Alenia Space, Prancis tahun 2020 itu berteknologi Very High Throughput Satellite (VHTS) berkapasitas 150 gigabita per detik (Gbps) dengan frekuensi Ka-Band. Tidak heran satelit SATRIA-1 dinobatkan sebagai satelit terbesar di Asia dan kelima di dunia.

```python

import folium
from folium import plugins


# menentukan kordinat Indonesia
map = folium.Map(location=[-2.4833826, 117.8902853], zoom_start=5)


# sisipkan gambar satelit pada satu folder

icon_plane = folium.features.CustomIcon('Satelite.png', icon_size=(100, 70))

marker = folium.Marker(location=[-2.4833826, 117.8902853], icon=icon_plane, tooltip='Satelite SATRIA-1 Indonesia, Simulation By ARI')


# menentukan titik point ke pulau
folium.Marker([-6.939008, 107.7407532], icon=folium.Icon(icon='signal', color='blue'), popup='Signal Pulau Jawa').add_to(map)
folium.Marker([0.133333, 117.5000003], icon=folium.Icon(icon='signal', color='blue'), popup='Signal Pulau Kalimantan').add_to(map)
folium.Marker([-0.942942, 100.371857], icon=folium.Icon(icon='signal', color='blue'), popup='Signal Pulau Sumatra').add_to(map)
folium.Marker([-0.89819, 119.86468], icon=folium.Icon(icon='signal', color='blue'), popup='Signal Pulau Sulawesi').add_to(map)
folium.Marker([-8.67324, 115.20338], icon=folium.Icon(icon='signal', color='blue'), popup='Signal Pulau Bali').add_to(map)
folium.Marker([-5, 140], icon=folium.Icon(icon='signal', color='blue'), popup='Signal Pulau Papua').add_to(map)



map.add_child(marker)


# RUTE SIGNAL SATELIT KE PULAU

route = folium.PolyLine(locations=[[-2.4833826, 117.8902853], [-6.939008, 107.7407532]], color='Blue')
route = folium.PolyLine(locations=[[-2.4833826, 117.8902853], [0.133333, 117.5000003]], color='Blue')
route = folium.PolyLine(locations=[[-2.4833826, 117.8902853], [-0.942942, 100.371857]], color='Blue')
route = folium.PolyLine(locations=[[-2.4833826, 117.8902853], [-0.89819, 119.86468]], color='Blue')
route = folium.PolyLine(locations=[[-2.4833826, 117.8902853], [-8.67324, 115.20338]], color='Blue')
route = folium.PolyLine(locations=[[-2.4833826, 117.8902853], [-5, 140]], color='Blue')




map.add_child(route)


# membuat animasi sinyal ke titik point

plugins.MarkerCluster().add_to(map)


plugins.AntPath(locations=[[-2.4833826, 117.8902853], [-6.939008, 107.7407532]]).add_to(map)
plugins.AntPath(locations=[[-2.4833826, 117.8902853], [0.133333, 117.5000003]]).add_to(map)
plugins.AntPath(locations=[[-2.4833826, 117.8902853], [-0.942942, 100.371857]]).add_to(map)
plugins.AntPath(locations=[[-2.4833826, 117.8902853], [-0.89819, 119.86468]]).add_to(map)
plugins.AntPath(locations=[[-2.4833826, 117.8902853], [-8.67324, 115.20338]]).add_to(map)
plugins.AntPath(locations=[[-2.4833826, 117.8902853], [-5, 140]]).add_to(map)

map
```
