```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data penduduk di Provinsi Jawa Barat
tahun = [2015, 2016, 2017, 2018, 2019, 2020]
jumlah_penduduk = [46797523, 47536889, 48298087, 49077034, 49898282, 50764992]

# Membuat plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Membuat bar 3D
dx = [1] * len(tahun)
dy = [1] * len(tahun)
dz = [0] * len(tahun)
dz = jumlah_penduduk

ax.bar3d(tahun, dy, dz, dx, dy, dz, color='skyblue')

# Konfigurasi plot
ax.set_xlabel('Tahun')
ax.set_ylabel('Dummy')
ax.set_zlabel('Jumlah Penduduk')
ax.set_title('Grafik Bar 3D Penduduk di Provinsi Jawa Barat')

# Tampilkan grafik bar 3D
plt.show()
```
