# **Visualisasi Pegunungan Dan Perairan  Menggunakan Matlab (Matrix Labolaory)**

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('3D Volcano Matlab')
ax.set_xlabel('Nilai X')
ax.set_ylabel('Nilai Y')
ax.set_zlabel('Nilai z')
plt.show()
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Generate data for the mountain
X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)
Z_mountain = np.sin(np.sqrt(X**2 + Y**2))

# Generate data for the sea
Z_sea = np.zeros_like(Z_mountain)

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Grafik Pegunungan
ax.plot_surface(X, Y, Z_mountain, cmap='terrain')

# Grafik Perairan
ax.plot_surface(X, Y, Z_sea, color='blue', alpha=0.5)

# Penamaan Label
ax.set_xlabel('Nilai X')
ax.set_ylabel('Nilai Y')
ax.set_zlabel('Nilai Z')
ax.set_title('3D Grafik Pegunungan dan Perairan (MATLAB)')

# Tampilkan Hasil
plt.show()
```
