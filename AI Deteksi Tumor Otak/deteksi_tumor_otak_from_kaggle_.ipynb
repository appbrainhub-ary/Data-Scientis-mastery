```python
import numpy as np
import pandas as pd
import os

import matplotlib.pyplot as plt
from IPython.display import display, Image
import cv2
```

# Exploratory data analysis

```python
HOME = '/kaggle/input/brain-mri-images-for-brain-tumor-detection'
```

```python
# get lists of images filenames
no_tumor_filenames = os.listdir(os.path.join(HOME, 'no'))
tumor_filenames = os.listdir(os.path.join(HOME, 'yes'))

# form array of Mat objects of yes/no tumor img
no_tumor_imgs = [cv2.imread(f"{HOME}/no/{filename}") for filename in no_tumor_filenames]
tumor_imgs = [cv2.imread(f"{HOME}/yes/{filename}") for filename in tumor_filenames]
```

```python
# display examples of imgs with NO tumor
example_n = 3 # number of images displayed
fig, axes = plt.subplots(1,example_n)
for i in range(example_n):
    axes[i].imshow(no_tumor_imgs[i])
axes[example_n//2].set_title('No tumor images')
None
```

```python
# display examples of imgs with tumor
example_n = 5 # number of images displayed
fig, axes = plt.subplots(1,example_n, figsize=(12,8))
for i in range(example_n):
    axes[i].imshow(tumor_imgs[i])
axes[example_n//2].set_title('With tumor images')
None
```

```python
# check the proportion between amount of images with and without tumor
fig, ax = plt.subplots()

categories = ['no', 'yes']
counts = [len(no_tumor_filenames), len(tumor_filenames)]
bar_labels = ['red', 'blue']
bar_colors = ['tab:red', 'tab:blue']

ax.bar(categories, counts, label=bar_labels, color=bar_colors)

ax.set_title('Tumor vs no tumor imgs')
```
